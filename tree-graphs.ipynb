```python
class TreeNode:
  def __init__(self, val, left=None, right=None):
    self.val = val
    self.left = left
    self.right = right

  def __str__(self):
    return str(self.val)
```

```python
A = TreeNode(1)
B = TreeNode(2)
C = TreeNode(3)
D = TreeNode(4)
E = TreeNode(5)
F = TreeNode(10)

A.left = B
A.right = C
B.left = D
B.right = E
C.right = F
print(A)
```

```python
#DFS(preorder; node,left,right)

def dfs(Node):
  if Node is None:
    return False
  print(Node)
  dfs(Node.left)
  dfs(Node.right)


dfs(A)
```

```python
def dfs_in(node):
  if node is None:
    return

  dfs_in(node.left)
  print(node)
  dfs(node.right)
dfs_in(A)
```

```python
def dfs_p(node):
  if node is None:
    return
  dfs_p(node.left)
  dfs_p(node.right)
  print(node)

dfs_p(A)
```

```python
def pre(node):
  stk=[]
  height=0
  stk.append(node)
  while(stk):
    node = stk.pop()
    print(node)
    if node.right: stk.append(node.right)
    height+=1
    # print(height)
  return height
pre(A)
```

```python
from collections import deque
def bfs(node):
  q = deque()
  q.append(node)
  while q:
    node = q.popleft()
    if node.left: q.append(node.left)
    if node.right: q.append(node.right)
    print(node)
bfs(A)
```

```python
def bst(node, target):
  if node is None:
    return False
  if node.val == target:
    return True
  if node.val>target: return bst(node.left, target)
  if node.val<target: return bst(node.right, target)

print(bst(A, 10))
```

```python
# Adjacency matrix
M=[]
for i in range(ng):
  M.append([0]*ng)
M
```

```python
for i,j in g:
  M[i][j] = 1
M
```

```python
# if its undirected
MM=[]
for i in range(ng):
  MM.append([0]*ng)
for i,j in g:
  MM[i][j] = 1
  MM[j][i] = 1
MM
```

```python
#Adjacency list
from collections import defaultdict
D=defaultdict(list)
for i,j in g:
  D[i].append(j)
D
```
